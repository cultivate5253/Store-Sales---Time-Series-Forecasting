```python
#データ量が多いので、減らす3054348件→55242
df=df.loc['2017-08-01':]
df=df.round(2)
```

```python
len(df)
```

```python
test_days=2
test_ind=test_days*1782
test_ind
```

```python
train=df.iloc[:-test_ind]
test=df.iloc[-test_ind:]
```

```python
train
```

```python
test
```

データのスケーリング


```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(train)
```

```python
scaled_train=scaler.transform(train)
scaled_test=scaler.transform(test)

```

時系列ジェネレーター

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
length=1782
batch_size=1
generator=TimeseriesGenerator(scaled_train,scaled_train,length=length,batch_size=batch_size)
```

```python
len(scaled_train),len(scaled_test)
```

```python
X,y=generator[0]
```

```python
#flatten():多次元配列を一次元配列に変換することができます。
print(f"下表を考慮して : \n {X.flatten()}")
print(f'これは、yが何であるかを予測するものである。 : \n {y}')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
```

```python
scaled_train.shape
```

モデルの作成

```python
# モデルを定義する
model = Sequential()

# LSTM層
model.add(LSTM(100,input_shape=(length,scaled_train.shape[1])))

# 最終予測（各特徴ごとに1ニューロン）
model.add(Dense(scaled_train.shape[1]))

model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

アーリーストッピング

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(monitor='val_loss',patience=1)
validation_generator = TimeseriesGenerator(scaled_test,scaled_test, 
                                           length=length, batch_size=batch_size)
```

```python
model.fit(generator,epochs=5,
                    validation_data=validation_generator,
                   callbacks=[early_stop])
```

```python
model.history.history.keys()
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
```

では、このロジックをforループに組み込んで、テスト範囲全体の未来を予測することにしよう。

注：ここで注意するのは、出力とその寸法です。独自のprint()ステートメントを追加して、実際に何が起こるかを見てみましょう。

```python
n_features = scaled_train.shape[1]
test_predictions = []

first_eval_batch = scaled_train[-length:]
current_batch = first_eval_batch.reshape((1, length, n_features))

for i in range(len(test)):
    
    # 1つ先のタイムスタンプの予測値を取得（[0]で[配列]の代わりに数字だけを入力する場合）
    current_pred = model.predict(current_batch)[0]
    
    # 予測を保存する
    test_predictions.append(current_pred) 
    
    # バッチを更新して予測値を含めるようにし、最初の値を削除する
    current_batch = np.append(current_batch[:,1:,:],[[current_pred]],axis=1)
```
